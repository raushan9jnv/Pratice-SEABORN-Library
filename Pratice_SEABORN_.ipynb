```python
import seaborn as sns
import pandas as pd
```

```python
sales= pd.read_excel('/content/drive/MyDrive/Data Science/Datasets/sales.xlsx')
```

```python
sales.head()
```

```python
sales.nunique(axis=0)
```

```python
sales.shape
```

```python
sales.columns
```

```python
sns.relplot('Sales','Profit',data=sales)
```

```python
sns.relplot('Sales','Profit',data=sales,hue='Order Priority',style='Ship Mode')
```

```python
sns.relplot('Sales','Profit',data=sales,size='Discount',sizes=(25,100))
```

```python
sns.relplot('Order Date','Sales',data=sales,kind='line')
```

```python
sales['year']=sales['Order Date'].dt.year
```

```python
sales.year.unique()
```

```python
sns.relplot('year','Sales',data=sales,kind='line')
```

```python
sns.relplot('year','Sales',data=sales,kind='line',ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',ci='sd') #confidencial interval   standard deviation
```

```python
sns.relplot('year','Sales',data=sales,kind='line',estimator=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode')
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',style= "Product Category")
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',style= "Product Category",ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col= "Product Category",ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col= "Region",ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col= "Region",ci=None,col_wrap= 4)
```
