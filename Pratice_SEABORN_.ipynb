```python
import seaborn as sns
import pandas as pd
```

```python
sales= pd.read_excel('/content/drive/MyDrive/Data Science/Datasets/sales.xlsx')
```

```python
sales.head()
```

```python
sales.nunique(axis=0)
```

```python
sns.relplot('Sales','Profit',data=sales)
```

```python
sns.relplot('Sales','Profit',data=sales,hue='Order Priority',style='Ship Mode')
```

```python
sns.relplot('Sales','Profit',data=sales,size='Discount',sizes=(25,100))
```

```python
sns.relplot('Order Date','Sales',data=sales,kind='line')
```

```python
sales['year']=sales['Order Date'].dt.year
```

```python
sales.year.unique()
```

```python
sns.relplot('year','Sales',data=sales,kind='line')
```

```python
sns.relplot('year','Sales',data=sales,kind='line',ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',ci='sd') #confidencial interval   standard deviation
```

```python
sns.relplot('year','Sales',data=sales,kind='line',estimator=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode')
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',style= "Product Category")
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',style= "Product Category",ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col= "Product Category",ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col= "Region",ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col= "Region",ci=None,col_wrap= 4)
```

```python
sns.relplot('year','Sales',data=sales)
```

```python
sns.relplot('Sales','Profit',data=sales,col='Customer Segment',col_wrap=2)
```

```python
sns.relplot('Sales','Profit',data=sales,col='Customer Segment',col_wrap=2,hue="Order Priority")
```

```python
sns.relplot('Sales','Profit',data=sales,col='Customer Segment',col_wrap=2,hue="Ship Mode")
```

```python
sns.relplot('Sales','Profit',data=sales,col='Customer Segment',row= 'Product Category',hue="Ship Mode")
```

```python
sns.catplot("Ship Mode","Sales",data=sales)
```

```python
sns.catplot("Ship Mode","Sales",data=sales,jitter=False)
```

```python
sns.catplot("Ship Mode","Sales",data=sales.query("Profit > 5000"),jitter=False)
```

```python
sns.catplot("Ship Mode","Sales",data=sales.query("Profit > 1000"),kind='swarm')
```

#Box Plot chart

```python
sns.catplot(x="Product Category",y="Shipping Cost",data=sales,kind='box')
```

```python
sns.catplot("Product Category","Shipping Cost",hue='Ship Mode',data=sales,kind='box')
```

```python
sns.catplot("Product Category","Shipping Cost",hue='Ship Mode',hue_order=["Delivery Truck",'Regular Air','Express Air'],data=sales,kind='box')
```

```python
sns.catplot("Product Category","Shipping Cost",hue='Ship Mode',col='Ship Mode',row='Customer Segment',data=sales,kind='box')
```

#Violin Plot

```python
sns.catplot(x="Product Category",y="Shipping Cost",data=sales,kind='violin')
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",data=sales,hue='Ship Mode',kind='violin')
```

```python
sns.catplot(x="Product Category",y="Shipping Cost",data=sales,hue='Ship Mode',kind='violin',inner='stick')
```
